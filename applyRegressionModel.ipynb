# 1. Import necessary libraries

```python

#library for EDA
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")

```

# 2. Loading dataset 

```python
# read the cvs file
#df = pd.read_csv("CarPrice_data1.csv")
df = pd.read_csv("CarPrice_main_dataset.csv")
df.info()
```

# Converting datatypes

```python
df.info()
```

```python
#converting string object data type to category data type
df['CarCompany'] = df['CarCompany'].astype('category')
df['EngineType'] = df['EngineType'].astype('category')
df['Transmission'] = df['Transmission'].astype('category')
df['RegisteredCIty'] = df['RegisteredCIty'].astype('category')
df['CarColor'] = df['CarColor'].astype('category')
df['BodyType'] = df['BodyType'].astype('category')


#converting numeric object data type to float data type
df['EngineCapacity(cc)'] = df['EngineCapacity(cc)'].astype('float')
df.info()
```

```python
Continuous_features =[ 'ModelYear', 'Mileage(km)','EngineCapacity(cc)',  'Price(PKR_lacs)']
Categorical_features = ['CarCompany','EngineType', 'Transmission', 'RegisteredCIty', 'CarColor', 'BodyType']
```

```python
Selected_features=['ModelYear',
 'EngineCapacity(cc)',
 'CarCompany',
 'EngineType',
 'Transmission',
 'RegisteredCIty',
 'CarColor',
 'BodyType']
```

```python
Selected_columns =  Selected_features + ['Price(PKR_lacs)']

Final_data_for_ML1 = df[Selected_columns]
Final_data_for_ML1.head()
```

# Data Transforms - Encoding


```python
Final_data_for_ML1 = pd.get_dummies(Final_data_for_ML1, 
                                   columns=Categorical_features)
Final_data_for_ML1.head()
```

# Model Building

#### Steps
1.   Seperating data into features and target.
2.   Train test split.
3.   Scaling of train data
4.   Model 1 -- Linear Regression
5.   Model 2 -- polynomial Regression
6.   Model 3 -- SVR Regression

## step 1: Seperating data into input features and target feature

```python
features = Final_data_for_ML1.drop(['Price(PKR_lacs)'], axis = 1)
target = Final_data_for_ML1 ['Price(PKR_lacs)']
features.head()
#target.head()
```

## step 2: Train-test-split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size = 0.3, random_state = 100)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

## step 3: Scaling of train data

```python
from sklearn.preprocessing import MinMaxScaler
Scaler = MinMaxScaler()
X_train = Scaler.fit_transform(X_train)
X_test = Scaler.transform(X_test)

X_train_scaled = pd.DataFrame(X_train, columns = features.columns)
X_train_scaled.head()
```

## step 4: Apply Model 1 -- Linear Regression

```python
# Default Linear Regression model without any tuning - base metric
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

Linear_Regression_default = LinearRegression()
Linear_Regression_default.fit(X_train, y_train)

Predictions_train_d = Linear_Regression_default.predict(X_train)
Predictions_test_d =Linear_Regression_default.predict(X_test)

print('Linear Regression default Model parameters:', Linear_Regression_default.get_params())
print('--------------------------------------------------')
print('\nLinear Regression default intercept :', Linear_Regression_default.intercept_)
print('\nLinear Regression default coefficients : ')
print('\n')
Coeficients = pd.DataFrame(Linear_Regression_default.coef_, index = features.columns, columns=['Coefficients'])
print(Coeficients)
print('--------------------------------------------------')
print('\n')
df1 = pd.DataFrame({'Model': ['Linear_Regression_default', 'Linear_Regression_default'],
                    'Data': ['Train', 'Test'], 
                    'MSE': [mean_squared_error(y_train, Predictions_train_d), mean_squared_error(y_test, Predictions_test_d)],
                    'RMSE': [np.sqrt(mean_squared_error(y_train, Predictions_train_d)), np.sqrt(mean_squared_error(y_test, Predictions_test_d))],
                    'MAE': [mean_absolute_error(y_train, Predictions_train_d), mean_absolute_error(y_test, Predictions_test_d)],
                   'R^2': [r2_score(y_train, Predictions_train_d), r2_score(y_test, Predictions_test_d)]})
df1
```

## step 6: Apply Model 3 -- Ridge_Regression

```python
# Default Ridge Regression model without any tuning - base metric
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

Ridge_Regression_default = Ridge()
Ridge_Regression_default.fit(X_train, y_train)
Predictions_train_d = Ridge_Regression_default.predict(X_train)
Predictions_test_d =Ridge_Regression_default.predict(X_test)

print('Ridge Regression default Model parameters:', Ridge_Regression_default.get_params())
print('--------------------------------------------------')
print('\nRidge Regression default intercept :', Ridge_Regression_default.intercept_)
print('\nRidge Regression default coefficients : ')
print('\n')
Coeficients = pd.DataFrame(Ridge_Regression_default.coef_, index = features.columns, columns=['Coefficients'])
print(Coeficients)
print('--------------------------------------------------')
print('\n')
df3 = pd.DataFrame({'Model': ['Ridge_Regression_default', 'Ridge_Regression_default'],
                    'Data': ['Train', 'Test'], 
                    'MSE': [mean_squared_error(y_train, Predictions_train_d), mean_squared_error(y_test, Predictions_test_d)],
                    'RMSE': [np.sqrt(mean_squared_error(y_train, Predictions_train_d)), np.sqrt(mean_squared_error(y_test, Predictions_test_d))],
                    'MAE': [mean_absolute_error(y_train, Predictions_train_d), mean_absolute_error(y_test, Predictions_test_d)],
                    'R^2': [r2_score(y_train, Predictions_train_d), r2_score(y_test, Predictions_test_d)]})
df3
```

## step 7: Apply Model 4 -- SVR default Regression

```python
from sklearn.svm import SVR
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

SVR_default = SVR()#(kernel='linear')
SVR_default.fit(X_train, y_train)
Predictions_train_d = SVR_default.predict(X_train)
Predictions_test_d =SVR_default.predict(X_test)

print('SVR default Model parameters:', SVR_default.get_params())
print('--------------------------------------------------')
print('\n')
df4 = pd.DataFrame({'Model': ['SVR_default', 'SVR_default'],
                    'Data': ['Train', 'Test'], 
                    'MSE': [mean_squared_error(y_train, Predictions_train_d), mean_squared_error(y_test, Predictions_test_d)],
                    'RMSE': [np.sqrt(mean_squared_error(y_train, Predictions_train_d)), np.sqrt(mean_squared_error(y_test, Predictions_test_d))],
                    'MAE': [mean_absolute_error(y_train, Predictions_train_d), mean_absolute_error(y_test, Predictions_test_d)],
                   'R^2': [r2_score(y_train, Predictions_train_d), r2_score(y_test, Predictions_test_d)]})
df4
```

## collective result


```python
collective_result = pd.concat([df1,df3,df4], ignore_index = True)
collective_result
```

# 9. Train the model using 100% data set available

## 9.1 Train linear regrssion model using 100% data available

```python
# input features and target
#print(features.shape)
#print(target.shape)

Linear_Regression = LinearRegression()
Linear_Regression.fit(features, target)
Predictions = Linear_Regression.predict(features)

print('Linear Regression Model parameters:', Linear_Regression.get_params())
print('\nLinear Regression intercept :', Linear_Regression.intercept_)
print('\nLinear Regression coefficients : ')
print('\n')
Coeficients = pd.DataFrame(Linear_Regression.coef_, index = features.columns, columns=['Coefficients'])
print(Coeficients)
print('\n')
print('-------------------------------------------------------')
print('Performance of SVR default Model')
final_LR = pd.DataFrame(({'Model': ['Linear_Regression'],
                          'MSE': [mean_squared_error(target, Predictions)],
                    'RMSE': [np.sqrt(mean_squared_error(target, Predictions))],
                    'MAE': [ mean_absolute_error(target, Predictions)],
                   'R^2': [r2_score(target, Predictions)]}))
final_LR
```

## 9.2. Train  Ridge Regression model using 100% data available

```python
# input features and target
#print(features.shape)
#print(target.shape)

Ridge_Regression = Ridge()
Ridge_Regression.fit(features, target)
Predictions = Ridge_Regression.predict(features)

print('Ridge Regression Model parameters:', Ridge_Regression.get_params())
print('\nRidge Regression intercept :', Ridge_Regression.intercept_)
print('\nRidge Regression coefficients : ')
print('\n')
Coeficients = pd.DataFrame(Ridge_Regression.coef_, index = features.columns, columns=['Coefficients'])
print(Coeficients)
print('\n')
print('-------------------------------------------------------')
print('Performance of Ridge Regression')
final_RR = pd.DataFrame(({'Model': ['Ridge_Regression'],
                          'MSE': [mean_squared_error(target, Predictions)],
                    'RMSE': [np.sqrt(mean_squared_error(target, Predictions))],
                    'MAE': [ mean_absolute_error(target, Predictions)],
                   'R^2': [r2_score(target, Predictions)]}))
final_RR
```

## 9.3 Train the SVR_default model using 100% data available

```python
# input features and target
#print(features.shape)
#print(target.shape)

Support_Vector_Regression = SVR()#(kernel='linear')
Support_Vector_Regression.fit(features, target)
Predictions = Support_Vector_Regression.predict(features)

print('Support Vector Regression (SVR) Model parameters:', Support_Vector_Regression.get_params())
print('\n Support Vector Regression (SVR) intercept :', Support_Vector_Regression.intercept_)
print('\n')
print('-------------------------------------------------------')
print('Performance of Support Vector Regression (SVR)  Model')
final_SVR = pd.DataFrame(({'Model': ['Support_Vector_Regression(SVR)'],
                     'MSE': [mean_squared_error(target, Predictions)],
                    'RMSE': [np.sqrt(mean_squared_error(target, Predictions))],
                    'MAE': [ mean_absolute_error(target, Predictions)],
                   'R^2': [r2_score(target, Predictions)]}))
final_SVR
```

```python
combineResultfor100 = pd.concat([final_LR,final_RR,final_SVR], ignore_index = True)
combineResultfor100
```
